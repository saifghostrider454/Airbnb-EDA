#### Lets Handle Outliers, If any

````python
num_cols = df.select_dtypes(exclude=object).columns[2:]

fig, ax = plt.subplots(1, len(num_cols), figsize=[28, 10])

for i in range(len(num_cols)):
    sns.boxplot(df[num_cols[i]], ax=ax[i])
````

````python
class HandlingOutliers:
    
    def __init__(self, input_df):
        self.output_df = input_df.copy()
        
    def fit(self, columns=None, **fit_params):
        return self
    
    def transform(self, columns=None, **transform_params):
        if columns is None:
            raise ValueError("Please specify columns for Handling outlier.")

        for col in columns:
            Q1 = self.output_df[col].quantile(0.25)
            Q3 = self.output_df[col].quantile(0.75)

            IQR = Q3 - Q1

            minimum_threshold = Q1 - 1.5 * IQR
            maximum_threshold = Q3 + 1.5 * IQR

            # Handling outliers
            self.output_df[col] = np.where(self.output_df[col] > maximum_threshold, maximum_threshold,
                                            np.where(self.output_df[col] < minimum_threshold, minimum_threshold,
                                                     self.output_df[col]))
        
        return self.output_df
````

````python
ho = HandlingOutliers(df)

df = ho.transform(num_cols)
````

````python
# Chart - 9 visualization code

numeric_columns = df.select_dtypes(exclude='object').columns
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')


# Compute correlations
correlations = df[numeric_columns].corrwith(df['price'])

# Define a custom colormap based on correlation values
norm = plt.Normalize(correlations.min(), correlations.max())
colors = plt.cm.viridis(norm(correlations))

# Plot correlation with response variable with gradient color
plt.figure(figsize=(10, 8))
bar_plot = correlations.plot.barh(title='Correlation with Response Variable', 
                                  fontsize=15, grid=True, color=colors)
bar_plot.invert_yaxis()  # Invert y-axis to have highest correlations at the top
plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap='viridis'), ax=bar_plot)
plt.show()
````
