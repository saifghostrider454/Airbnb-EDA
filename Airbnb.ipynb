## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Making a copy of my dataset
df = dataframe.copy()
````

````python
df.drop(columns=['id', 'host_id'], inplace=True)
````

#### Handling Missing values

````python
df.isnull().sum().to_frame(name='count').style.background_gradient('GnBu')
````

````python
df.head(3)
````

````python
df['last_review'] = df.groupby(['neighbourhood_group', 'neighbourhood', 'room_type'])['last_review'].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else pd.NaT))
````

````python
df = df[~df['last_review'].isna()]
````

````python
df.isnull().sum().to_frame(name='count').style.background_gradient('GnBu')
````

````python
df['reviews_per_month'] = df.groupby('last_review')['reviews_per_month'].transform(lambda x: x.fillna(x.mean()))
````

````python
df.isnull().sum().to_frame(name='count').style.background_gradient('GnBu')
````

````python
df.dropna(inplace=True)
````

````python
df.isnull().sum().to_frame(name='count').style.background_gradient('GnBu')
````

#### Lets Handle Outliers, If any

````python
num_cols = df.select_dtypes(exclude=object).columns[2:]

fig, ax = plt.subplots(1, len(num_cols), figsize=[28, 10])

for i in range(len(num_cols)):
    sns.boxplot(df[num_cols[i]], ax=ax[i])
````

````python
class HandlingOutliers:
    
    def __init__(self, input_df):
        self.output_df = input_df.copy()
        
    def fit(self, columns=None, **fit_params):
        return self
    
    def transform(self, columns=None, **transform_params):
        if columns is None:
            raise ValueError("Please specify columns for Handling outlier.")

        for col in columns:
            Q1 = self.output_df[col].quantile(0.25)
            Q3 = self.output_df[col].quantile(0.75)

            IQR = Q3 - Q1

            minimum_threshold = Q1 - 1.5 * IQR
            maximum_threshold = Q3 + 1.5 * IQR

            # Handling outliers
            self.output_df[col] = np.where(self.output_df[col] > maximum_threshold, maximum_threshold,
                                            np.where(self.output_df[col] < minimum_threshold, minimum_threshold,
                                                     self.output_df[col]))
        
        return self.output_df
````

````python
ho = HandlingOutliers(df)

df = ho.transform(num_cols)
````

````python
num_cols = df.select_dtypes(exclude=object).columns[2:]

fig, ax = plt.subplots(1, len(num_cols), figsize=[28, 10])

for i in range(len(num_cols)):
    sns.boxplot(df[num_cols[i]], ax=ax[i])
````

#### Chart - 1

````python
# Chart - 1 visualization code

plt.figure(figsize=[29, 6])

top_5_hotels = df['name'].value_counts().head(5)

# Define a color palette
colors = sns.color_palette("husl", len(top_5_hotels))

# Plotting the bar plot with customized colors
ax = sns.barplot(x=top_5_hotels.index, y=top_5_hotels.values, palette=colors)

plt.xticks(rotation=45)
plt.xlabel('Hotel Name')  # Adding x-label
plt.title('Top 5 Hotels by Count')  # Adding plot title

# Adding text annotations on top of each bar
for i, count in enumerate(top_5_hotels.values):
    ax.text(i, count, str(count), ha='center', va='bottom')

# Adding hotel names at the center of the bars
for i, name in enumerate(top_5_hotels.index):
    bar_width = ax.patches[i].get_width()
    bar_center = ax.patches[i].get_x() + bar_width / 2
    ax.text(bar_center, top_5_hotels[i] / 2, name, ha='center', va='center', rotation=90)

# Removing the x and y axes
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)

plt.show()
````

#### Chart - 2

````python
# Chart - 2 visualization code
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']  # You can add more colors if needed

# Grouping by 'neighbourhood_group' and getting counts
group_counts = df.groupby('neighbourhood_group')['neighbourhood_group'].count()

# Calculate total count
total_count = group_counts.sum()

# Plotting the outer pie chart with custom colors and formatted labels
plt.figure(figsize=(8, 8))
patches, _ = plt.pie(group_counts, labels=[f'{label} ({count} - {count/total_count*100:.1f}%)' for label, count in zip(group_counts.index, group_counts)], startangle=140, pctdistance=0.85, colors=colors)

# Drawing a white circle at the center to create a hole
center_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
plt.title('Neighbourhood Groups Distribution (Donut Chart)')
plt.show()
````

##### 1. Why did you pick the specific chart?

The specific chart chosen here is a donut chart. Donut charts are suitable for displaying the distribution of categorical data, such as the distribution of different categories within a single variable. In this case, the data represents the distribution of listings among different neighborhood groups in New York City.

Donut charts are effective in conveying proportions while also allowing for a clearer visualization of the total count. By having the data displayed both as slices and as a whole, viewers can easily understand the contribution of each category to the overall dataset.

Additionally, the inclusion of annotations with arrows pointing to each section of the donut chart enhances the clarity by explicitly connecting each section to its label, making it easier for viewers to interpret the data.

Overall, the donut chart provides a visually appealing and informative way to present the distribution of neighborhood groups in this dataset.

````python
# Chart - 3 visualization code

scaler = StandardScaler()

scaled_features = scaler.fit_transform(df[['price', 'minimum_nights', 'number_of_reviews']])

# Convert the scaled features array to a DataFrame
scaled_df = pd.DataFrame(scaled_features, columns=['price', 'minimum_nights', 'number_of_reviews'])

plt.figure(figsize=(10, 6))
sns.kdeplot(scaled_df.price, label='Price', shade=True)
sns.kdeplot(scaled_df.minimum_nights, label='Minimum Nights', shade=True)
sns.kdeplot(scaled_df.number_of_reviews, label='MNumber of Reviews', shade=True)

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Kernel Density Estimate (KDE) Plot for Price and Minimum Nights and number_of_reviews')
plt.legend()

# Display the plot
plt.show()
````
