## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
cols = dataframe.select_dtypes(exclude='object').columns[4:]

# Calculate the number of numerical columns
col_range = len(cols)

# Set up the subplots based on the number of numerical columns
fig, ax = plt.subplots(1, col_range, figsize=(25, 6))

# Loop through each numerical column and create a boxplot
for i in range(col_range):
    sns.boxenplot(y=dataframe[cols[i]], ax=ax[i])
    ax[i].set_title(cols[i])  # Set the title for each subplot

plt.tight_layout()
plt.show()
````

````python
class OutliersRemover:
    
    def __init__(self, input_df):
        self.output_df = input_df.copy()
        
    def fit(self, columns=None, **fit_params):
        return self
    
    def transform(self, columns=None, **transform_params):
        if columns is None:
            raise ValueError("Please specify columns for outlier removal.")

        for col in columns:
            Q1 = self.output_df[col].quantile(0.25)
            Q3 = self.output_df[col].quantile(0.75)

            IQR = Q3 - Q1

            minimum_threshold = Q1 - 1.5 * IQR
            maximum_threshold = Q3 + 1.5 * IQR

            # Remove outliers
            self.output_df = self.output_df[~((self.output_df[col] < minimum_threshold) | (self.output_df[col] > maximum_threshold))]
        
        return self.output_df
````

````python
ors = OutliersRemover(dataframe)

ndf = ors.transform(cols)
````

````python
# Write your code to make your dataset analysis ready.
cols = ndf.select_dtypes(exclude='object').columns[4:]

# Calculate the number of numerical columns
col_range = len(cols)

# Set up the subplots based on the number of numerical columns
fig, ax = plt.subplots(1, col_range, figsize=(25, 6))

# Loop through each numerical column and create a boxplot
for i in range(col_range):
    sns.boxenplot(y=ndf[cols[i]], ax=ax[i])
    ax[i].set_title(cols[i])  # Set the title for each subplot

plt.tight_layout()
plt.show()
````
